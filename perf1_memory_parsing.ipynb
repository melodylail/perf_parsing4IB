```python
from matplotlib.ticker import MultipleLocator, FormatStrFormatter
sns.set()


for val in sorted(service_name_set):
    compute_business_job_1sec = df[df['Series'] == val]
    compute_business_job_1sec.columns = ['Series','Time','Value']
    test = compute_business_job_1sec[['Value']]
    print(compute_business_job_1sec[['Series']])
    y_cur = []
    ymajorLocator   = MultipleLocator(100) #y0.5
    ymajorFormatter = FormatStrFormatter('%d') #y
    yminorLocator   = MultipleLocator(100) #y0.1

    ax = plt.subplot(111) 

    ax.yaxis.set_major_locator(ymajorLocator)
    ax.yaxis.set_major_formatter(ymajorFormatter)

    for index, row in test.iterrows():
        val_list = row.str.split('\t')
        #print(index)
        print(val_list[-1][-1])
        y_cur.append(int(val_list[-1][-1]))
    y_cur_Series = pd.Series(y_cur)
    percentile_90 = y_cur_Series.astype(float).quantile(.9)

    print(percentile_90)

    plt.plot(x, y_cur_Series, marker='o', color='mediumvioletred')
    plt.title(val)
    plt.axhline(y=percentile_90, color='g', linestyle='-')
    plt.show()
```
